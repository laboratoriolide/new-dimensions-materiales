## Librerías
En este caso podemos usar únicamente las librería `pandas`.

```python
import pandas as pd
```

Podemos acceder a los datos usando la función `pd.read_csv()` y el URL de los datos.

```python
felicidad = pd.read_csv('https://raw.githubusercontent.com/cienciadedatos/datos-de-miercoles/master/datos/2019/2019-08-07/felicidad.csv')
felicidad.head()
```

Podemos usar las funciones `min()` y `max()` para obtener los valores mínimos y máximos de la columna `anio`.

```python
anio_min = felicidad['anio'].min()
anio_max = felicidad['anio'].max()
print(anio_min, anio_max)
```

Ahora podemos agrupar los datos por año y calcular la media y desviación estándar de la columna `puntaje_felicidad`.

```python
prom_felicidad = felicidad.groupby('anio').escalera_vida.mean()
std_felicidad = felicidad.groupby('anio').escalera_vida.std()
felicidad_global = pd.DataFrame({'promedio': prom_felicidad, 'desviacion_standard': std_felicidad})
felicidad_global
```

```python
felicidad.
```

Ahora podemos usar la función `loc()` para obtener los datos de los años de nuestro interés.

```python
felicidad_global = felicidad_global.loc[[anio_min, anio_max], :]
felicidad_global
```

Para acceder a los datos de países filtrar nuestros datos tal como los hemos visto en clases.

```python
felicidad[felicidad.anio == 2006].pais
```

```python
felicidad_dos = felicidad[(felicidad.pais == 'Brasil')  | (felicidad.pais == 'Australia')]
felicidad_dos.head()
```

Luego podemos usar la función `loc()` para obtener los datos de los años de nuestro interés.

```python
felicidad_dos = felicidad_dos.loc[(felicidad_dos.anio == anio_min) | (felicidad_dos.anio == anio_max), ['pais', 'anio', 'escalera_vida']]
felicidad_dos
```
