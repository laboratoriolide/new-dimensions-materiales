```python
import requests
import re
from io import StringIO
import pandas as pd
```

```python
#Conexion a sitio web de datos abiertos con requests
con = requests.get('https://www.datosabiertos.gob.ec/dataset/')
```

```python
con
```

```python
#Revisamos contenido de la conexion
con.text
```

```python
#Dividimos el contenido de la conexion por saltos de linea
for t in con.text.split('\n'):
    print(t)
```

```python
#Buscamos organizaciones en el contenido de la conexion
#Como ejemplo vamos a revisar si la organizacion galapagos esta en el contenido de la conexion
org = []
for j in con.text.split('\n'):
    if 'organiza' and 'galapagos' in j.lower():
        org.append(j)

org
```

```python
#Extraemos el link de la organizacion
[terminacion] = re.findall('href="(.*)" ', org[0])
terminacion
```

```python
#Creamos un nuevo vinculo de busqueda
url_gal = 'https://www.datosabiertos.gob.ec' + terminacion
url_gal
```

```python
#Conectamos a la nueva url
con_gal = requests.get(url_gal)
con_gal.text
```

```python
#Dividimos el contenido de la conexion por saltos de linea
data_url = []
for t in con_gal.text.split('\n'):
    if 'dataset' in t.lower() and 'esterilizacion' in t.lower() and 'agencia' not in t.lower():
        data_url.append(t)
        print(t)
```

```python
[data_est] = re.findall('href="(.+)" ', data_url[0])
data_est
```

```python
url = 'https://www.datosabiertos.gob.ec' + data_est
url
```

```python
esterilizacion = requests.get(url)
esterilizacion.text
```

```python
descarga = []
for t in esterilizacion.text.split('\n'):
    if 'csv' in t.lower():
        descarga.append(t)
```

```python
descarga
```

```python
url_descarga = 'https://www.datosabiertos.gob.ec'+ re.findall('href="(.*)" ', descarga[1])[0]
url_descarga = url_descarga.replace('" title="', '/download/')
url_descarga
```

```python
requests.get(url_descarga).text
```

```python
StringIO(requests.get(url_descarga).text)
```

```python
datos_ester = pd.read_table(StringIO(requests.get(url_descarga).text), sep=';')
datos_ester.head()
```

```python
datos_ester.to_csv('datos_esterilizacion.csv', index=False)
```
